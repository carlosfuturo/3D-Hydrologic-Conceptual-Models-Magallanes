# Magallanes — Visualizador de Modelos 3D

Este notebook muestra los modelos 3D del proyecto hidrogeológico Magallanes (IDIEM).  
Cada modelo es un Plotly interactivo cargado desde el portal ArcGIS Online.

| Cuenca | Descripción |
|--------|-------------|
| **Penitente** | DEM, pozos, niveles freáticos, lagunas |
| **El Oro** | DEM, pozos, niveles freáticos, lagunas |
| **Robalo** | DEM, pozos, niveles freáticos, lagunas |
| **Combinado** | Las 3 cuencas en una sola vista |

> **Instrucciones:** Ejecutar la celda siguiente y seleccionar el modelo en el desplegable.

```python

import threading
import http.server
import time
from pathlib import Path
from IPython.display import display, HTML

HTML_DIR = Path(r"\\Mac\Home\Documents\Proyectos\Flores\IDIEM\01 Magallanes\02 PyScripts Magallanes\Figuras_3D_Leapfrog")
PORT = 8766  # nuevo puerto — evita caché del navegador

ITEMS = {
    "Modelo Combinado (3 cuencas)": "3D_Leapfrog_COMBO_3Cuencas.html",
    "Cuenca Penitente":             "3D_Leapfrog_Penitente.html",
    "Cuenca El Oro":                "3D_Leapfrog_El_Oro.html",
    "Cuenca Robalo":                "3D_Leapfrog_Robalo.html",
}

# Verificar archivos
print("Verificando archivos:")
for titulo, archivo in ITEMS.items():
    existe = (HTML_DIR / archivo).exists()
    print(f"  {'✓' if existe else '✗'} {titulo}")

# Servidor con cabeceras no-cache
class _Handler(http.server.SimpleHTTPRequestHandler):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, directory=str(HTML_DIR), **kwargs)
    def end_headers(self):
        self.send_header("Cache-Control", "no-store, no-cache, must-revalidate")
        self.send_header("Pragma", "no-cache")
        self.send_header("Expires", "0")
        super().end_headers()
    def log_message(self, *args):
        pass

try:
    _srv = http.server.HTTPServer(("localhost", PORT), _Handler)
    threading.Thread(target=_srv.serve_forever, daemon=True).start()
    time.sleep(0.5)
    print(f"✓ Servidor iniciado en http://localhost:{PORT}")
except OSError:
    print(f"✓ Servidor ya activo en http://localhost:{PORT}")

# Cache-buster en la URL (timestamp)
ts = int(time.time())
options_html = "\n".join(
    f'<option value="{archivo}?v={ts}">{titulo}</option>'
    for titulo, archivo in ITEMS.items()
)
default = f"{list(ITEMS.values())[0]}?v={ts}"

display(HTML(f"""
<h3>Seleccionar modelo 3D:</h3>
<select id="mdl-select"
  style="width:380px;padding:8px;font-size:14px;margin-bottom:10px;"
  onchange="document.getElementById('mdl-frame').src='http://localhost:{PORT}/'+this.value">
{options_html}
</select>
<br>
<iframe id="mdl-frame"
  src="http://localhost:{PORT}/{default}"
  width="100%" height="800px" frameborder="0"
  style="border:1px solid #ccc;background:white;">
</iframe>
"""))
```
